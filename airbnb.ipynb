# handling missing values

```python
air["last review"] = pd.to_datetime(air["last review"], errors="coerce")
```

```python
air.info()
```

```python
air.fillna({
    "reviews per month": 0, 
    "last review": air["last review"].min()  # Note: air, not df
}, inplace=True)
```

```python
air.dropna(subset=["NAME","host name"],inplace=True)
```

```python
#drop the columns
air=air.drop(columns=["license","house_rules"],errors="ignore")
air
```

```python
#remove dollar sign and conver to float
air["price"] = air["price"].replace('[\$,]', '', regex=True).astype(float)
air["service fee"] = air["service fee"].replace('[\$,]', '', regex=True).astype(float)
```

```python
air.head()
```

remove duplicates

```python
air.drop_duplicates(inplace=True)
```

```python
air.info()
```

```python
air.describe()
```

visualization 

What is the distributing of listing prices?

```python
plt.figure(figsize=(10,6))
sns.histplot(air["price"],bins=50,kde=True,color="red")
plt.title("Distribution of listing price")
plt.xlabel("Price $")
plt.ylabel("Frequency")
plt.show()
```

How are different room types distributed ?

```python
air.columns
```

```python
air["room type"]
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="room type",data=air,color="hotpink")
plt.title("Room Type Distribution")
plt.xlabel("room type ")
plt.ylabel(" count")
plt.show()
```

How are listing distributed across different neighbourshoods?

```python
plt.figure(figsize=(12,8))
sns.countplot(y="neighbourhood group",data=air,color="lightgreen",order=air["neighbourhood group"].value_counts().index)
plt.title("Number of Listings by Neighboorhood Group")
plt.xlabel("count")
plt.ylabel("Neighborhood Group")
plt.show()
```

what is Relationship between price and room type

```python
plt.figure(figsize=(10,6))
sns.boxplot(x="room type",y="price",data=air,palette="Set1")
plt.xlabel("Room Type")
plt.ylabel("Price $")
plt.legend(title="Room Type")
plt.show()
```

how has the number of review chnage over with time

```python
air['last review'] = pd.to_datetime(air['last review'])

reviews_over_time = air.groupby(air['last review'].dt.to_period('M')).size()

plt.figure(figsize=(12, 6))
reviews_over_time.plot(kind='line', color='red')

plt.title('Number of Reviews Over Time')
plt.xlabel("Years")
plt.ylabel("Number of Reviews")
plt.show()
```
